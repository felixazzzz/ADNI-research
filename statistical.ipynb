```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('ALL_Axial_T2_STAR_11_20_2023.csv')
df
```

```python
# convert the "sex" column into binary numeric values
df['Sex'] = df['Sex'].map({'M': 0, 'F': 1})
```

```python
# give histogram of the data based on these columns age
df.hist(column='Age', bins=20, grid=False, figsize=(12,8), color='#86bf91', zorder=2, rwidth=0.9)
plt.title('Age Distribution of the Patients')
plt.xlabel('Age')
plt.ylabel('Number of Patients')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df is your DataFrame and it already exists with a 'Sex' column where 0 is Male and 1 is Female

# Count the occurrence of each class
sex_count = df['Sex'].value_counts()

# Labels for the classes
labels = ['Male', 'Female']

# Colors for each section
colors = ['#ff9999','#66b3ff']

# Explode the 1st slice (i.e. 'Male')
explode = (0.1, 0)  

# Creating the pie chart
plt.pie(sex_count, labels=labels, colors=colors, explode=explode, autopct='%1.1f%%', startangle=140)

# Adding a title
plt.title('Distribution of Sex')

# Adding a legend
plt.legend(labels, title="Sex", loc="best")

# Show the plot
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df is your DataFrame and it has 'Sex' and 'Group' columns

# Define the figure for plotting
fig, axs = plt.subplots(1, 2, figsize=(14, 7))  # 1 row, 2 columns for Male and Female

# Define the labels and colors for each group (assuming group labels are numeric or can be categorized easily)
# If your groups are not numeric or need specific labeling, adjust the labels accordingly.
group_labels = df['Group'].unique()
group_labels.sort()  # Sort the labels if necessary
colors = plt.cm.tab20.colors  # Using tab20 colormap for distinct colors

# Loop through each sex category (0 for Male, 1 for Female)
for i, sex in enumerate(['Male', 'Female']):
    # Filter the DataFrame by sex
    df_sex = df[df['Sex'] == i]
    
    # Count the occurrence of each group within the filtered DataFrame
    group_count = df_sex['Group'].value_counts().sort_index()  # Sort by index if your data is not sorted
    
    # Create the pie chart for the current sex
    axs[i].pie(group_count, labels=group_labels, colors=colors[:len(group_labels)], autopct='%1.1f%%', startangle=140)
    axs[i].set_title(f'Distribution of Group within {sex}')
    
# Add a super title for the figure
plt.suptitle('Proportion of Each Label by Sex')

# Show the plot
plt.show()
```

```python
# for 'Acq Date' column, we will convert the date to a datetime type
# the date is in the format of 'mm/dd/yyyy'
df['Acq Date'] = pd.to_datetime(df['Acq Date'], format='%m/%d/%Y')
df.hist(column='Acq Date', bins=20, grid=False, figsize=(12,8), color='#86bf91', zorder=2, rwidth=0.9)
plt.title('Aquired Date Distribution of the Patients')
plt.xlabel('Aquired Date')
plt.ylabel('Number of Patients')
plt.show()
```
